```python
import pandas as pd 
import numpy as np

import datetime
from datetime import date
from dateutil.relativedelta import relativedelta


length_of_timeframe = 5  #CHANGE BACK TO 1 IF FAILS
#5 years is the maximum available from Tiingo


today = date.today() - datetime.timedelta(days = 3) ##Delete the relativedelta subtraction when done
end = today.strftime("%Y-%m-%d") #Today's date in YYYY-MM-DD format

today = today - relativedelta(years=length_of_timeframe)
start = today.strftime("%Y-%m-%d") #Date that is 5 years before
```

```python
#CSV Citation: https://datahub.io/core/s-and-p-500-companies-financials#data

SP500_extract = pd.read_csv('/Users/kaylie/Downloads/SP500_Ticker_List.csv')

SP500_Company_Names = SP500_extract['Description']
SP500_Company_Names = SP500_Company_Names.astype(str)
SP500_Company_Names = SP500_Company_Names.to_numpy()

SP500_Tickers = SP500_extract['Symbol']
SP500_Tickers = SP500_Tickers.astype(str)
SP500_Tickers = SP500_Tickers.to_numpy()
SP500_Tickers = SP500_Tickers.tolist()
```

```python
import multiprocessing as mp
from multiprocessing import pool
import time

import import_ipynb

from ipynb.fs.full.Tiingo_Retrieval_Function import *

split_arrays_of_tickers = np.array_split(SP500_Tickers, 24)

reformatted_splits = []
for list in split_arrays_of_tickers:
    newly_formatted = list.tolist()
    reformatted_splits.append(newly_formatted)


start_time = time.time()
stock_price_data = pd.DataFrame()

for i in range(1,5):
    pool_main = mp.Pool(processes = 6)
    extracted_data = []
    arguments = []
    
    for a in range (6*(i-1), 6*i):
        sub_argument = (start, end, reformatted_splits[a])
        arguments.append(sub_argument)
        
    extracted_data = pool_main.starmap(extract_hourly_timeseries_price_data, arguments)
       
    for a in range(0, 6):
        df1 = extracted_data[a]
        stock_price_data = pd.concat([df1, stock_price_data], axis=1)
    
total_execution_time = (time.time() - start_time)

print(total_execution_time)
```
