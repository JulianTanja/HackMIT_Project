```python
#23,500
#divide out of percentage of max
#multiple weights by total emissions for each type of gas
#based off of potency 
#CO2
#Methane
#Nitros 

'''
global warming potential (100yr)
co2 = 1
"times co2"

SF6 (23.5k times worse than CO2) – 23,500  (23500/23500)
HFC – 1,300                                (1300/23500)
PFC – 6630                                 (6630/23500)
NF3 – 16,100                               (16100/23500)
Methane (80 times worse than CO2) – 25     (25/23500)
C02 – 1                                    (1/23500)
Nitrous Oxide – 265                        (265/23500)

'''
def scale(df):
    
    #change to float, so we can run calculations 
    df["CO2 emissions (non-biogenic) "] = df["CO2 emissions (non-biogenic) "].str.replace(',', '')
    df["CO2 emissions (non-biogenic) "] = df["CO2 emissions (non-biogenic) "].astype(float).fillna(0)
    
    df["Methane (CH4) emissions "] = df["Methane (CH4) emissions "].str.replace(',', '')
    df["Methane (CH4) emissions "] = df["Methane (CH4) emissions "].astype(float).fillna(0)
    
    df["Nitrous Oxide (N2O) emissions "] = df["Nitrous Oxide (N2O) emissions "].str.replace(',', '')
    df["Nitrous Oxide (N2O) emissions "] = df["Nitrous Oxide (N2O) emissions "].astype(float).fillna(0)
    
    df["HFC emissions"] = df["HFC emissions"].str.replace(',', '')
    df["HFC emissions"] = df["HFC emissions"].astype(float).fillna(0)
    
    df["PFC emissions"] = df["PFC emissions"].str.replace(',', '')
    df["PFC emissions"] = df["PFC emissions"].astype(float).fillna(0)
    
    df["SF6 emissions "] = df["SF6 emissions "].str.replace(',', '')
    df["SF6 emissions "] = df["SF6 emissions "].astype(float).fillna(0)
    
    df["NF3 emissions"] = df["NF3 emissions"].str.replace(',', '')
    df["NF3 emissions"] = df["NF3 emissions"].astype(float).fillna(0)
    
    df["Total reported direct emissions"] = df["Total reported direct emissions"].str.replace(',', '')
    df["Total reported direct emissions"] = df["Total reported direct emissions"].astype(float).fillna(0)

    #scaling factor * og num 
    df["CO2 emissions (non-biogenic) "] = df["CO2 emissions (non-biogenic) "] * (1/23500)#treat like dictionary/ updating keys, but actually w dataframe
    df["Methane (CH4) emissions "] = df["Methane (CH4) emissions "] * (25/23500)
    df["Nitrous Oxide (N2O) emissions "] = df["Nitrous Oxide (N2O) emissions "] * (265/23500)
    df["HFC emissions"] = df["HFC emissions"]* (1300/23500)
    df["PFC emissions"] = df["PFC emissions"] * (6630/23500)
    df["SF6 emissions "] = df["SF6 emissions "] * (23500/23500)
    df["NF3 emissions"] = df["NF3 emissions"] * (16100/23500)
    
```

```python
scale(F122017)
scale(F122018)
scale(F122019)
scale(F122020)
```

```python
'''
add all * total emissions = adjusted
call adjusted total emissions

'''
def adjTotEmi(df):
    sum = 0
    sum =  (df["CO2 emissions (non-biogenic) "] + df["Methane (CH4) emissions "] 
    + df["Nitrous Oxide (N2O) emissions "] + df["HFC emissions"] + df["PFC emissions"] + df["SF6 emissions "] + df["NF3 emissions"]) 
    df["Adjusted direct emissions"] = sum * (df["Total reported direct emissions"])
    
```
